```python
import pandas as pd;

df = pd.read_csv("customer_shopping_behavior.csv");
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
# Fill the null values in Review Rating with median value of each category. 
#For clothing category null value, we put clothing category media value.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# For tracking the column easily, I make all the column string lower case.
df.columns = df.columns.str.lower()
df.columns
```

```python
# For tracking the column easily, I replaces all the spaces in the column name with '_'.
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
df.columns
```

```python
# Create a new column named aged_group. And split the ages into 4 groups. 
labels = ['Young Adult', 'Adult', 'Middle-age', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
df[['age_group','age']].head(10)
```

```python
df.frequency_of_purchases.unique()
```

```python
# Create a new column purchase_frequency_days, because the frequency of purchases are written in text.
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Annually': 365,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90,
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
# Let's see if discount_applied and promo_code_used column have the same value or not.
#df[['discount_applied', 'promo_code_used']].head(10)
(df.discount_applied == df.promo_code_used).all()
```

```python
# As discount_applied and promo_code_used have same values, let's remove one of them.
df = df.drop('promo_code_used', axis=1)
df.columns
```
